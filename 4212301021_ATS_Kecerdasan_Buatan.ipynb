```python
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# Fungsi Keanggotaan Triangular
# ==============================
def triangular(x, a, b, c):
    if x <= a or x >= c:
        return 0
    elif a < x < b:
        return (x - a) / (b - a)
    elif b <= x < c:
        return (c - x) / (c - b)
    else:
        return 0

# ==============================
# Fuzzifikasi
# ==============================
def fuzzifikasi(tinggi, angin):

    # Tinggi Air
    surut = triangular(tinggi, 0, 0, 80)
    normal = triangular(tinggi, 60, 100, 140)
    pasang = triangular(tinggi, 120, 160, 200)

    # Angin
    tenang = triangular(angin, 0, 0, 5)
    sedang = triangular(angin, 3, 6.5, 10)
    kencang = triangular(angin, 8, 11.5, 15)

    return surut, normal, pasang, tenang, sedang, kencang

# ==============================
# Inferensi (Rule Base Sederhana)
# ==============================
def inferensi(surut, normal, pasang, tenang, sedang, kencang):

    # Rule:
    # Surut & Tenang → Aman
    aman = min(surut, tenang)

    # Normal & Sedang → Waspada
    waspada = min(normal, sedang)

    # Pasang atau Kencang → Bahaya
    bahaya = max(pasang, kencang)

    return aman, waspada, bahaya

# ==============================
# Defuzzifikasi (Metode Max Membership)
# ==============================
def defuzzifikasi(aman, waspada, bahaya):

    nilai = {
        "Aman": aman,
        "Waspada": waspada,
        "Bahaya": bahaya
    }

    return max(nilai, key=nilai.get)

# ==============================
# Input
# ==============================
tinggi_input = float(input("Masukkan Ketinggian Air Laut (cm): "))
angin_input = float(input("Masukkan Kecepatan Angin (m/s): "))

# Proses Fuzzy
s, n, p, t, sd, k = fuzzifikasi(tinggi_input, angin_input)
a, w, b = inferensi(s, n, p, t, sd, k)
hasil = defuzzifikasi(a, w, b)

# ==============================
# Output
# ==============================
print("\nDerajat Keanggotaan:")
print(f"Surut   : {s:.2f}")
print(f"Normal  : {n:.2f}")
print(f"Pasang  : {p:.2f}")
print(f"Tenang  : {t:.2f}")
print(f"Sedang  : {sd:.2f}")
print(f"Kencang : {k:.2f}")

print("\nStatus Laut:", hasil)
```
